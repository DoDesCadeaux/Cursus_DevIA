### Univariate content analysis

```python
target = 'class'
```

```python
data[target].unique()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Countplot for target')

ax = sns.countplot(data=data, x=target)

for container in ax.containers:
    ax.bar_label(container)

plt.show()

print(data[target].value_counts())
```

```python
data[[target]].describe()
```

```python
feat_cols = data.columns.to_list()
feat_cols.remove(target)
feat_cols
```

```python
import numpy as np

N_cat = len(feat_cols)

plt.figure(figsize=(10, 5*N_cat))

for i in np.arange(N_cat):
    col = feat_cols[i]
    plt.subplot(N_cat, 1, i+1)
    plt.title(f"kdeplot of col={col}")

    ax = sns.countplot(data=data, x=col)

    for container in ax.containers:
        ax.bar_label(container)

plt.tight_layout()
```

```python
N_cat = len(feat_cols)

plt.figure(figsize=(10, 5*N_cat))

for i in np.arange(N_cat):
    col = feat_cols[i]
    plt.subplot(N_cat, 1, i+1)
    plt.title(f"kdeplot of col={col}")

    ax = sns.countplot(data=data, x=col, hue=target)
    for container in ax.containers:
        ax.bar_label(container)

plt.tight_layout()
```

```python
GB1 = data[[target, 'gill-size']].groupby(['gill-size', target]).agg({target:'count'})

GB1.columns = ['count']
GB1.reset_index(inplace=True)
display(GB1)

GB1_gill_size_b = GB1[GB1['gill-size'] == 'b']
GB1_gill_size_n = GB1[GB1['gill-size'] == 'n']

plt.subplot(1, 2, 1)
plt.title('Broad gill size')
plt.pie(x=GB1_gill_size_b['count'], labels=GB1_gill_size_b['class'], autopct='%1.1f%%')

plt.subplot(1, 2, 2)
plt.title('Narrow gill size')
plt.pie(x=GB1_gill_size_n['count'], labels=GB1_gill_size_n['class'], autopct='%1.1f%%')
plt.show()
```
