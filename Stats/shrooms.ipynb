```python
GB1 = data[[target, 'gill-size']].groupby(['gill-size', target]).agg({target:'count'})

GB1.columns = ['count']
GB1.reset_index(inplace=True)
display(GB1)

GB1_gill_size_b = GB1[GB1['gill-size'] == 'b']
GB1_gill_size_n = GB1[GB1['gill-size'] == 'n']

plt.subplot(1, 2, 1)
plt.title('Broad gill size')
plt.pie(x=GB1_gill_size_b['count'], labels=GB1_gill_size_b['class'], autopct='%1.1f%%')

plt.subplot(1, 2, 2)
plt.title('Narrow gill size')
plt.pie(x=GB1_gill_size_n['count'], labels=GB1_gill_size_n['class'], autopct='%1.1f%%')
plt.show()
```

```python
import pandas as pd

crosstab = pd.crosstab(data['class'], data['gill-size'], normalize='columns')
display(crosstab)

sns.heatmap(crosstab, annot=True, cmap='Blues', cbar=False)
plt.show()
```

```python
data_encoded = pd.get_dummies(data)

corr_matrix = data_encoded.corr()
display(corr_matrix)
```

```python
plt.figure(figsize=(25, 20))
sns.heatmap(data=corr_matrix, annot=False, cmap='coolwarm', vmax=1, vmin=-1, fmt='.2f')
plt.show()
```

```python
threshold = 0.5
corr_target_e = corr_matrix["class_e"].drop(['class_e', 'class_p'])
corr_target_p = corr_matrix["class_p"].drop(['class_e', 'class_p'])

relevant_features_e = corr_target_e[corr_target_e.abs() > threshold].index.tolist()
relevant_features_p = corr_target_p[corr_target_p.abs() > threshold].index.tolist()

selected_columns = list(set(relevant_features_e + relevant_features_p))

selected_columns.extend(['class_e', 'class_p'])

print("Colonnes sélectionnées pour le pairplot :", selected_columns)
```

```python
data_subset = data_encoded[selected_columns]

subset_corr_matrix = data_subset.corr()

plt.figure(figsize=(14, 12))
sns.heatmap(subset_corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Heatmap des Colonnes Sélectionnées')
plt.show()
```
