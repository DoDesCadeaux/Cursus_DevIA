### 1. Faire un (line-)plot qui montre:

    - le nombre uniques de sports par année
    - placer des points comme marqueurs
    - changer le style de ligne en pointillé 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('jo-dataset.csv')
dataset.head(10)
```

```python
unique_sports_per_year = dataset['Sport'].groupby(dataset['Year']).unique()
sports_count_per_year = unique_sports_per_year.apply(len)
sports_count_per_year_df = pd.DataFrame(sports_count_per_year)
sports_count_per_year_df
```

```python
fig, ax = plt.subplots()

ax.plot(sports_count_per_year.index, sports_count_per_year, "o:")
plt.xlabel("Year")
plt.ylabel("Number of sports")
plt.title("Number of Sports per Year")
plt.show()
```

### 2. Faire un graphique composé qui montre:

    - le nombre uniques de sports par année (voir 1.)
    - avec une décomposition par genre homme-femme, indiqué dans une légende
    - changer les couleurs en 'cyan' et 'magenta'
    - appliquer un 'alpha' de '0.5'
    - changer la légende de position: en bas à gauche 

```python
men = dataset[dataset['Sex'] == "M"]
men_sport = men['Sport'].groupby(men['Year']).unique()
men_sport_count = men_sport.apply(len)
men_sport_count
```

```python
women = dataset[dataset['Sex'] == 'F']
women_sport = women['Sport'].groupby(women['Year']).unique()
women_sport_count = women_sport.apply(len)
women_sport_count
```

```python
plt.plot(men_sport_count.index, men_sport_count, "o-", c='c', alpha=0.5)
plt.plot(women_sport_count.index, women_sport_count, "o-", c='m', alpha=0.5)
plt.legend(['Male', 'Female'], bbox_to_anchor=(0, 0))
plt.show()
```

### 3. Faire un bar-plot qui montre:

    - le top 10 des noms qui ont participés aux plus de sports différents,
    - trié par: nombre de sports différents, nombre de médailles
    - changer l'orientation des labels sur l'axe-x ('xticks')
    - changer le nom de l'axe-x: 'top 10' ('xlabel')

```python
sports_per_name = dataset.groupby('Name')['Sport'].nunique()
top_10_athletes_per_sport = sports_per_name.sort_values(ascending=False)
top_10_athletes_per_sport

medals_per_name = dataset.groupby('Name')['Medal'].count()
top_10_athletes_per_medals = medals_per_name.sort_values(ascending=False)
top_10_athletes_per_medals
```

```python
athlete_stats = pd.DataFrame({
    'Nr of sports': top_10_athletes_per_sport,
    'Nr of medals': top_10_athletes_per_medals
})

top_10_athletes = athlete_stats.sort_values(by=['Nr of sports', 'Nr of medals'], ascending=False).head(10)
top_10_athletes
```

```python
plt.figure(figsize=(12, 6))
top_10_athletes['Nr of sports'].plot(kind='bar')
plt.title('Top 10 Athletes by Number of Different Sports')
plt.xlabel('Top 10')
plt.ylabel('Number of Sports')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Faire un pie-plot qui montre les info de 'Carl Schuhmann':

    - le nombre de médailles (confondues) pour chaque sport
    - avec le type de sport pour lequel cet athlète a le plus de medailles, mis en évidence



```python
carl_schuhmann = dataset[dataset['Name'] == 'Carl Schuhmann']
carl_schuhmann_medals = carl_schuhmann.groupby(carl_schuhmann['Sport'])['Medal'].count()
carl_schuhmann_medals_not_zero = carl_schuhmann_medals[carl_schuhmann_medals != 0]
carl_schuhmann_medals_not_zero
```

```python
plt.pie(data=carl_schuhmann_medals, x=carl_schuhmann_medals_not_zero, labels=carl_schuhmann_medals_not_zero.index, autopct="%1.1f%%")
plt.title('Carl Schuhmann Medals')
plt.show()
```

### Faire un histogramme qui montre:

    - l'age
    - par tranche d'age de 3 années 
    - combien il y'a de personnes par tranches (chaque tranche de 3 ans correspond a 1 barre d'histogramme)

```python
age = dataset['Age'].dropna()

bins = range(int(age.min()), int(age.max()) + 3, 3)

plt.hist(age, bins=bins)
plt.show()
```

### Faire graphique composé d' histogrammes qui montre:

    - la taille (height)
    - par tranche de 4 cm
    - decomposé par homme-femme
    - avec un alpha = 0.6



```python
males = dataset[dataset['Sex'] == 'M']
males_height = males['Height'].dropna()

women = dataset[dataset['Sex'] == 'F']
women_height = women['Height'].dropna()

bins = range(int(dataset['Height'].min()), int(dataset['Height'].max() + 4), 4)

plt.hist(women_height, bins, alpha=0.6)
plt.hist(males_height, bins, alpha=0.6)
plt.show()
```
