```python
from sklearn.preprocessing import OrdinalEncoder

import pandas as pd
import numpy as np
```

```python
fav_colors = ['Red', 'Green', 'Blue']
gender = ['M', 'F']

data = pd.DataFrame({
    'fav_colors': np.random.choice(fav_colors, 500),
    'gender': np.random.choice(gender, 500)
})

data.head(10)
```

```python
x = data.to_numpy()
x
```

```python
encoder = OrdinalEncoder()
encoder.fit(x)
x_encoded = encoder.transform(x)
x_encoded
```

```python
n_samples, n_features = x.shape

x_encoded = np.empty((n_samples, n_features), dtype=np.float64)

categories = []
for idx_col in range(n_features):
    categories.append(np.unique(x[:, idx_col]))

for idx_col, col_category in enumerate(categories):
    for idx_cat, category in enumerate(col_category):
        print(idx_cat, category)
        x_encoded[x[:, idx_col] == category, idx_col] = idx_cat        

x_encoded
```
