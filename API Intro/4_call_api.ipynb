```python
url_communes = "https://www.odwb.be/api/explore/v2.1/catalog/datasets/code-postaux-belge/records?limit=20"

response = req.get(url_communes)

if response.status_code == 200:
    data_json = response.json()
    records = data_json['results']
else:
    print("Erreur lors de la récupération des données des communes")

communes = []
for record in records:
    fields = record
    commune = {
        'name': fields.get('column_2'),
        'latitude': fields.get('column_4'),
        'longitude': fields.get('column_3')
    }
    communes.append(commune)

df_communes = pd.DataFrame(communes)
print("Communes de Belgique:")
print(df_communes)
```

```python
def get_weather(lat, lon):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&hourly=temperature_2m"
    response = req.get(url)

    if response.status_code == 200:
        json_data = pd.json_normalize(response.json())
        weather_data = json_data[['latitude', 'longitude', 'hourly.time', 'hourly.temperature_2m']].explode(column=["hourly.time", "hourly.temperature_2m"])
        
        return weather_data
```

```python
weather_forecasts = []

for index, row in df_communes.iterrows():
    forecast = get_weather(row['latitude'], row['longitude'])
    if forecast is not None:
        forecast['commune'] = row['name']
        weather_forecasts.extend(forecast.to_dict('records'))

weather_forecasts_df = pd.DataFrame(weather_forecasts)
weather_forecasts_df
```
